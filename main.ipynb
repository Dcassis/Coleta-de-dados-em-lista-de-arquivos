```python
import pandas as pd
import os
```

```python
Diretorio = '/Users/dcassis/Documents/Pós_Graduação - Data Science and Learning Machine/Bootcamp - Arquiteto de Big Data/Projetos/coleta_de_dados/vendas_veiculos'
```

```python
lista_df = []
lista_path = os.listdir(Diretorio)
lista_path[:5]
```

```python
lista_df = []
for arquivo in lista_path:
    local_arquivo = os.path.join(Diretorio, arquivo)
    df = pd.read_csv(local_arquivo, sep=';', encoding='latin-1')
    lista_df.append(df)
df_vendas = pd.concat(lista_df)
df_vendas[:10]
```

```python
df_vendas.to_csv(Diretorio + 'vendas_por_data.csv', sep = ';', encoding='latin-1', index=False)
```
